## Recipe Changes

- 300g beef -> 400g beef (quantity arguement)
- pancetta -> bacon (price argument)
- 120ml white wine + 120 chicken broth -> 240 red wine (same as drinking) + half cube chicken broth cube (equivalent to 250ml broth). More interesting with more pronouced wine taste. Convenient with red wine, since you already drink it with Bolognese. Increase broth for more flavor.
- 300g crushed tomato -> 400g crushed tomato (quantity arguement)
- Taste with MSG (why not?!)





## Homemade Pasta

- 3 eggs 
- 300g tippo 0 flour
- Mix fast, volcano method in bowl.
- Rolled to 6 flatness, cut tagliatelle



## Reflections

Good Bolognese. Eyeballed salt in paste water. Should be 3%, quite high. Dish turned out overall pretty undersalted. Also forgot pepper...
Was also in a rush, so cooktime was a little short. Cooked for one hour, then 30 min after adding milk. Recipe calls for twice of this.

Consider parmesan in the bolognese itself. Bought a 125 stick for topping. Was left with about 40g after topping 4 servings. Might buy a larger stick, and save 80g for topping, and use rest in the bolognese (add with milk?).

Flatness 6 was a bit too think. Concider 7-8 next time.
