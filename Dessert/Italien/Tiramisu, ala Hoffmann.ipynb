# Tiramisu, ala Hoffmann

James Hoffmann's scientificially iterated Tiramisu, providing a simple recipe of few ingredients, focusing on bringing out the best of the coffee flavor.

I utilized 10 biscuits and 18g per double espresso. These values were utilizied in his included Tiramisu calculator to provide the recipe. 

Video: https://www.youtube.com/watch?v=o3AMALUTp8o

Calculator: https://www.jameshoffmann.co.uk/tiramisu-calculator

## Recipe Changes

Input 10 biscuits in the calculator.

- 3 yolks -> 4 first price eggs yolks (they are very small) 


## Reflections

The coffe used was 2:1 18g Kaffebrenneriet Koffeinfri Tolima, lightroast. The coffee was nice, and useful to have caffein free, as there is quite a lot of coffee in this dessert, and could be hard to enjoy in the evening. However, the lightroastedness of the coffee made it very light and fruity, which didn't go too well with the sweet biscuits and custard. For future, try medium roast/light espresso roast.

The biscuits was a bit oversaturated. 1sec is enough to soak each finger.

Lastly, the custard could have been a bit stiffer. I used a too large bowl, which made whipping it inefficient.
